```python
#Import dependency
import pandas as pd
import re
import numpy as np
```

```python
#Read in CSV
prod_pd = pd.read_csv("International_energy_production.csv")
prod_pd
```

```python
print(len(prod_pd))
prod_pd.drop([i for i in range(1161, 1394)], axis=0, inplace=True)
print(len(prod_pd))
```

```python
prod_pd = prod_pd.dropna(subset=['API'])
```

```python
split = prod_pd['API'].str.split("-").str[0]
split
prod_pd['Energy Type'] = split
```

```python
prod_pd
```

```python
prod_pd['Energy Type'] = prod_pd['Energy Type'].str.replace('INTL.4411', 'Coal')
prod_pd['Energy Type'] = prod_pd['Energy Type'].str.replace('INTL.4413', 'Natural Gas')
prod_pd['Energy Type'] = prod_pd['Energy Type'].str.replace('INTL.4415', 'Petroleum')
prod_pd['Energy Type'] = prod_pd['Energy Type'].str.replace('INTL.4419', 'Nuclear, Renewables and Other')
prod_pd['Energy Type'] = prod_pd['Energy Type'].str.replace('INTL.4417', 'Nuclear')
prod_pd['Energy Type'] = prod_pd['Energy Type'].str.replace('INTL.4418', 'Renewables')
prod_pd['Energy Type'] = prod_pd['Energy Type'].str.replace('INTL.44', 'Total Energy Production')
prod_pd
```

```python
prod_pd.dtypes
```

```python
clean_prod_df = prod_pd.replace('--', '0.0')
clean_prod_df = clean_prod_df.replace('ie', '0.0')
clean_prod_df
```

```python
filter_col = [col for col in clean_prod_df if col.startswith('1')]

for col in filter_col:
    clean_prod_df[col] = clean_prod_df[col].astype(float)
    
    
```

```python
filter_col = [col for col in clean_prod_df if col.startswith('2')]

for col in filter_col:
    clean_prod_df[col] = clean_prod_df[col].astype(float)
```

```python

clean_prod_df.dtypes
```

```python
clean_prod_df.rename(columns={'Unnamed: 1':"Country"}, inplace = True)
clean_prod_df.dtypes
```

```python
countries_df = clean_prod_df.groupby(['Country',"Energy Type"])
clean_df = countries_df.first()
clean_df.head()
```

```python
clean_df.to_csv("Clean International Energy Production.csv")
```

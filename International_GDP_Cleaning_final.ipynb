```python
import pandas as pd
```

```python
file = "InternationalGDPv1.xlsx"
```

```python
#Reading .xlsx file
df = pd.read_excel(file)
```

```python
#Dropping unnamed column
df2 = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df2.head()
```

```python
#Removing non numeric data
df3 = df2.drop(42, axis=0)
```

```python
df3.tail()
```

```python
#Preparing for float data type switch 
df4 = df3.replace(
    {'..': 0.0})
```

```python
df4.head()
```

```python
#Preparing for float data type switch
df5 = df4.fillna(0)
```

```python
df5.head()
```

```python
#Preparing for float data type switch
cols = df5.columns

for col in cols:
    df5[col]=df5[col].astype(str).str.replace(',','')
```

```python
df5.head()
```

```python
df6 = df5.astype(float)
```

```python
df6.dtypes
```

```python
df6.head()
```

```python
df6.to_csv('International_GDP_Clean.csv', index=False)
```

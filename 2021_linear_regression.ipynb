```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
renew1_df = pd.read_csv("2021_Renewable_GDP1.csv")
renew1_df.head
```

```python
plt.scatter(renew1_df['2021_gdp'], renew1_df['2021_renew'])
plt.xlabel('GDP')
plt.ylabel('Energy Production')
plt.show()
```

```python
X = renew1_df['2021_gdp'].values.reshape(-1, 1)
```

```python
X.shape
```

```python
y = renew1_df['2021_renew']
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```

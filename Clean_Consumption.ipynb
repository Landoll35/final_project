```python
#Import dependency
import pandas as pd
```

```python
#Read in CSV
df = pd.read_csv("Consumption.csv")
df
```

```python
df = df.reset_index(drop=False)
df
```

```python
split = df['level_0'].str.split("-").str[0]
split
df['Energy Type'] = split
```

```python
df
```

```python
df['Energy Type'] = df['Energy Type'].str.replace('INTL.4411', 'Coal')
df['Energy Type'] = df['Energy Type'].str.replace('INTL.4413', 'Natural Gas')
df['Energy Type'] = df['Energy Type'].str.replace('INTL.4415', 'Petroleum')
df['Energy Type'] = df['Energy Type'].str.replace('INTL.4419', 'Nuclear, Renewables and Other')
df['Energy Type'] = df['Energy Type'].str.replace('INTL.4417', 'Nuclear')
df['Energy Type'] = df['Energy Type'].str.replace('INTL.4418', 'Renewables')
df['Energy Type'] = df['Energy Type'].str.replace('INTL.44', 'Total Energy Production')
```

```python
df
```

```python
df = df.replace('--', '0.0')
df = df.replace('ie', '0.0')
df
```

```python
df['level_2'] = df['level_2'].astype(float)
```
